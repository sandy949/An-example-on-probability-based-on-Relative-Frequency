```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv("/content/vehicles.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.drop(labels=data.columns[0:4],axis=1,inplace=True)
```

```python
data.drop(labels="title_status",axis=1,inplace=True)
```

```python
data.drop(labels=data.columns[14:],axis=1,inplace=True)
```

```python
data.drop(labels="VIN",axis=1,inplace=True)
```

```python
data.shape
```

```python
data.columns
```

```python
data.dropna(axis=0,thresh=12,inplace=True)
```

```python
data.isna().sum()
```

```python
data["odometer"].fillna(value=data["odometer"].mean(),inplace=True)
```

```python
data["condition"].unique()
```

```python
data["condition"].value_counts()
```

```python
data["condition"].fillna(value=data["condition"].value_counts().index[data["condition"].value_counts().argmax()],
                         inplace=True)
```

```python
data["paint_color"].fillna(value=data["paint_color"].value_counts().index[data["paint_color"].value_counts().argmax()],
                           inplace=True)
```

```python
data["transmission"].fillna(value=data["transmission"].value_counts().index[data["transmission"].value_counts().argmax()],
                           inplace=True)
```

```python
crosstab_df_mdl_yr_mfc = pd.crosstab(data["model"],
 [data["year"],data["manufacturer"]],rownames=["model"],
                                     colnames=["year","manufacturer"])
```

```python
crosstab_df_mdl_yr_mfc
```

```python
mapping_dict = dict()

for single_col in crosstab_df_mdl_yr_mfc.columns:
  mapping_dict[single_col] = crosstab_df_mdl_yr_mfc[single_col].index[crosstab_df_mdl_yr_mfc[single_col].argmax()]
```

```python
mapping_dict
```

```python
data.isna().sum()
```

```python
for k in mapping_dict.keys():
  boolean_mask = (data["year"] == k[0]) & (data["manufacturer"] == k[1])
  data.loc[boolean_mask,"model"] = data.loc[boolean_mask,"model"].fillna(value=mapping_dict[k],inplace=False)
```

```python
data.isna().sum()
```

```python
data.head()
```

```python
crosstab_df_mfr_yr_mdl = pd.crosstab(data["manufacturer"],
 [data["year"],data["model"]],rownames=["manufacturer"],colnames=["year","model"])
```

```python
crosstab_df_mfr_yr_mdl
```

```python
mapping_dict_2 = dict()
mapping_dict_2 = crosstab_df_mfr_yr_mdl.idxmax().to_dict()

```

```python
mapping_dict_2
```

```python
for k in mapping_dict_2.keys():
    boolean_mask = (data["year"] == k[0]) & (data["model"] == k[1])
    data.loc[boolean_mask, "manufacturer"] = data.loc[boolean_mask, "manufacturer"].fillna(value=mapping_dict_2[k], inplace=False)
```

```python
data.isna().sum()
```

```python
crosstab_df_mfr_yr_trns_cyn = pd.crosstab(data["manufacturer"],
 [data["year"],data["transmission"],data["cylinders"]],rownames=["manufacturer"],colnames=["year","transmission","cylinders"])
```

```python
crosstab_df_mfr_yr_trns_cyn
```

```python
mapping_dict_3 = dict()
mapping_dict_3 = crosstab_df_mfr_yr_trns_cyn.idxmax().to_dict()

```

```python
mapping_dict_3
```

```python
for k in mapping_dict_3.keys():
    boolean_mask = (data["year"] == k[0]) & (data["transmission"] == k[1]) &(data["cylinders"])
    data.loc[boolean_mask, "manufacturer"] = data.loc[boolean_mask, "manufacturer"].fillna(value=mapping_dict_3[k], inplace=False)
```

```python
crosstab_df_cyn_trns_mfr_mdl
```

```python
mapping_dict_4 = dict()
mapping_dict_4 = crosstab_df_cyn_trns_mfr_mdl.idxmax().to_dict()

```

```python
mapping_dict_4
```

```python
for k in mapping_dict_4.keys():
    boolean_mask = (data["manufacturer"] == k[0]) & (data["transmission"] == k[1]) &(data["model"])
    data.loc[boolean_mask, "cylinders"] = data.loc[boolean_mask, "cylinders"].fillna(value=mapping_dict_4[k], inplace=False)
```

```python
data.isna().sum()
```

```python
crosstab_df_sz_yr_mfr_mdl
```
